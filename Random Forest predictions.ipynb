```python
import pandas as pd
import pickle
import joblib
```

We're loading our previously saved pickle model
And then loading again our csv, to try to predict random elements from it.

```python
model = joblib.load('rf_model.pkl')
spam_db = pd.read_csv("spambase.data")
spam_db
```

Let's drop the label column and pick 10 random mails.

```python
spam_db.drop(spam_db.columns[len(spam_db.columns)-1], axis=1, inplace=True)
spam_10 = spam_db.sample(n = 10)
```

The last thing we want to do now is to predict the status of our mails, are they spams ?

```python
predicts = model.predict(spam_10)
predicts = pd.DataFrame(predicts, columns = ['Mail'])
predicts = predicts.replace({1:'A spam', 0:'Not a spam'})
predicts
```
